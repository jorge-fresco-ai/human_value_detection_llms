```python
import os
import pandas as pd
```

```python
directorio = "../DATA/data_tfg"
lista_archivos = []
for carpeta in os.listdir(directorio):
    ruta_carpeta = os.path.join(directorio, carpeta)
    if os.path.isdir(ruta_carpeta):  # Verifica si es una carpeta
        for archivo in os.listdir(ruta_carpeta):
            ruta_archivo = os.path.join(ruta_carpeta, archivo)
            if os.path.isfile(ruta_archivo):  # Verifica si es un archivo
                if "english" in ruta_archivo:
                    print(f"'{ruta_carpeta}/{archivo}'")
                    lista_archivos.append(ruta_carpeta+"/"+archivo)
                    #print(f"pd.to_csv('{ruta_carpeta}/{archivo}.tsv', sep='')")
```

### Categorías y Subcategorías de Schwarz

### Crear categorías para cada Dataset

Sumamos los valores de cada subcategoría para calcular las categorías en train, test y validation.

```python
for col in labels_train.columns[2:]:
    if 'constrained' not in col:
        print(col[:-9])
```

**Openness to Change**
- Self-Direction (Thought)
- Self-Direction (Action)
- Stimulation
- Hedonism

**Self-Enhancement**
- Achievement
- Power (Dominance)
- Power (Resources)
- Face

**Conservation**
- Security (Personal)
- Security (Societal)
- Tradition
- Conformity (Rules)
- Conformity (Interpersonal)
- Humility

**Self-Transcendence**
- Benevolence (Dependability)
- Benevolence (Caring)
- Universalism (Concern)
- Universalism (Nature)
- Universalism (Tolerance)

```python
def create_col_categs(df):
    df['openness_to_change attained']    = df.loc[:, [col for i, col in enumerate(df.columns[2:10], 
                                                    start=2) if i % 2 == 0]].sum(axis=1)
    df['openness_to_change constrained'] = df.loc[:, [col for i, col in enumerate(df.columns[2:10], 
                                                    start=2) if i % 2 != 0]].sum(axis=1)

    df['self_enhancement attained']      = df.loc[:, [col for i, col in enumerate(df.columns[10:18], 
                                                    start=2) if i % 2 == 0]].sum(axis=1)
    df['self_enhancement constrained']   = df.loc[:, [col for i, col in enumerate(df.columns[10:18], 
                                                    start=2) if i % 2 != 0]].sum(axis=1)
    
    df['conservation attained']          = df.loc[:, [col for i, col in enumerate(df.columns[18:30], 
                                                    start=2) if i % 2 == 0]].sum(axis=1)
    df['conservation constrained']       = df.loc[:, [col for i, col in enumerate(df.columns[18:30], 
                                                    start=2) if i % 2 != 0]].sum(axis=1)
    
    df['self_transcendence attained']    = df.loc[:, [col for i, col in enumerate(df.columns[30:40], 
                                                    start=2) if i % 2 == 0]].sum(axis=1)
    df['self_transcendence constrained'] = df.loc[:, [col for i, col in enumerate(df.columns[30:40], 
                                                    start=2) if i % 2 != 0]].sum(axis=1)
    return df
```

```python
labels_train = create_col_categs(labels_train)
labels_val   = create_col_categs(labels_val)
labels_test  = create_col_categs(labels_test)
```
